```python
spark.conf.set("spark.sql.shuffle.partitions", "5")
```

```python
from pyspark.sql.functions import col, spark_partition_id

# generate sample data with 50 records
data = [[str(i), f"name_{i}", f"company_{(i % 5) + 1}"] for i in range(1, 51)]

# specify column names
columns = ['Employee ID', 'Employee NAME', 'Company Name']

# creating a dataframe from the lists of data
dataframe = spark.createDataFrame(data, columns)

# display dataframe
display(dataframe)
```

```python
last=dataframe.collect()[-1]

last_id=last[0]
#dataframe.filter(dataframe['Employee ID']!=last_id).show()


df=dataframe.orderBy(col("Employee ID").desc(),col("Employee NAME").desc())
df.show()
#df=dataframe.filter(dataframe['Employee ID']!=last_id)
#df.show()


```
